## 方程式の問題


$x$,$y$,$z$を正の数とします。

$xy+z=43$, $yz+x=61$,$zx+y=47$を満たすとき, $x$,$y$,$z$の値を求めよ。

wolfram engine で解いてみます。


```
Solve[{x y+z==43,y z+x==61,z x+y==47},{x,y,z}]
```

５組あるようですが，よくわかりません。$(x,y,z)=(5,7,8)$は整数解のようです。近似値で解を求めてみます。

```
NSolve[{x y+z==43,y z+x==61,z x+y==47},{x,y,z}]//TeX
```

実数の範囲で５組あるようです。今回の正の数の組は４組です。

\begin{align*}
(x,y,z)=&(60.4651,0.698489,0.765756)\\
&(0.910101,45.8061,1.31183)\\
&(1.10009,1.44656,41.4087)\\
&(−6.47532,−7.95115,−8.48624)\\
&(5,7,8)
\end{align*}

グレブナー基底で確認してみます。

```
GroebnerBasis[{x y+z-43,y z+x-61,z x+y-47},{x,y,z}]
```

5次方程式で解けるようです。$z$についての５次方程式

$$z^5-43z^4-2z^3+2953z^2-5929z+2824=0$$

となっているので，これを調べると，

```
Factor[z^5-43z^4-2z^3+2953z^2-5929z+2824]
```

因数分解できます。

$$(z-8)(z^4-35z^3-282z^2+697z-353)=0$$

一応，方程式を解いて，グラフを見ておきます。

```
NSolve[z^5-43z^4-2z^3+2953z^2-5929z+2824==0,z]
```

```
Plot[z^5-43z^4-2z^3+2953z^2-5929z+2824,{z,-10,50}]
```
